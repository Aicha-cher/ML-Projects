# Matplotlib Examples

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.arange(1,3)
plt.plot(y,'<-.')
plt.plot(y+1,'b--')
plt.plot(y+2,'k:')
plt.show()
```

```python
plt.plot(y, color='blue', linestyle='-.', linewidth=4,marker='o', 
         markerfacecolor='red', markeredgecolor='black', 
         markeredgewidth=3, markersize=17);
plt.show()
```

```python
y = np.random.randn(1000)
plt.hist(y,25)
plt.grid()
plt.show()
```

```python
x = np.arange(0,4,0.2)
y = np.exp(-x)
e1 = 0.1 * np.abs(np.random.randn(len(y)))
plt.errorbar(x,y,yerr=e1, fmt='.-')
plt.show()
```

```python
plt.bar([1,2,3],[3,2,5])
plt.show()
```

```python
x = [45,35,20]
plt.figure(figsize=(4,4))
labls = ['Cats', 'Dogs' , 'Fishes']
plt.pie(x ,labels = labls)
plt.show()
```

```python
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x,y)
plt.show()
```

```python
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x,y)
plt.show()
```

```python
size = 50 * np.random.randn(1000)
color = np.random.randn(1000)
plt.scatter(x,y , c = color , s = size)
plt.show()
```

```python
plt.scatter(x,y , marker = 'x')
plt.show()
```

```python
theta = np.arange(0.,2.,1./180) * np.pi
plt.polar(3*theta,theta/5)
plt.polar(theta,np.cos(4*theta/5))
plt.show()
```

```python
x = np.arange(0, 2*np.pi, .01)
y = np.sin(x)
plt.plot(x, y)
plt.text(0.1 , 0.72 , 'Hello world')
plt.show()
```

```python
y = [13, 11, 13, 12, 13, 10, 30, 12, 11, 13, 12, 12, 12, 11,12]
plt.plot(y)
plt.ylim(ymax=35)
plt.annotate('this spot must really\nmean something', 
             xy=(6, 30), xytext=(8, 31.5), arrowprops=dict(facecolor='red',shrink=0.04))
plt.show()
```

```python
x = np.arange(0,10,0.1)
y = np.random.randn(len(x))
plt.plot(x,y)
plt.title(r'$\sin(2\pi x)\exp(-x)$ with the two asymptotes$\pm\exp(-x)$');
plt.show()
```

 # Subplotting

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(211)
ax1.plot([1, 2, 3], [1, 2, 3])
ax2 = fig.add_subplot(212)
ax2.plot([1, 2, 3], [3, 2, 1])
plt.show()
```
